```python
import numpy as np
df=pd.DataFrame({'score':[10,30,40,70,90]})
df['result']=np.where(df['score']>70,'pass','fail')
df['grade']=np.where(df['score']>70,'A',np.where(df['score']>40,'B','C'))
print(df)
```

```python
from sklearn.model_selection import train_test_split
df=pd.DataFrame({'hours':[1,2,3,4,5],'score':[70,55,65,85,90]})
x=df[['hours']]
y=df[['score']]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
print(x_train)
print(x_test)
```
