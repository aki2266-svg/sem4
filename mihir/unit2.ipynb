```python
import pandas as pd
import numpy as np
import seaborn as srs

f=pd.DataFrame({'Maths':[80,85,90,75,85],'Science':[78,82,88,80,82],'English':[70,75,80,72,79]})
srs.heatmap(f.corr(),annot=True,cmap='Blues',linecolor='Black',linewidth=2)
```

```python
import pandas as py 
import numpy as np
d=pd.DataFrame({'values':[0,40,35,32,45,100]})
print(d)
q1=d['values'].quantile(0.25)
print(q1)
```

```python
q3=d['values'].quantile(0.75)
print(q3)
```

```python
IQR=q3-q1
```

```python
print(IQR)
```

```python
lb=q1-1.5*IQR
print(lb)
```

```python
hb=q3+1.5*IQR
print(hb)
```

```python
d[(d['values']<lb) | (d['values']>hb)]
```

```python
clean_df=d[(d['values']>lb)&(d['values']<hb)]
print(clean_df)
```

```python
srs.boxplot(data=d['values'])
```

```python
g=pd.DataFrame({'featureA':[10,20,15,30,45],'featureB':[42,35,22,35,40],'featureC':[52,65,70,75,80],'gender':['m','f','m','f','m'],'category':['typeX','typeY','typeX','typeY','typeX']})
pd.plotting.parallel_coordinates(g,'category',cols=['featureA','featureB','featureC'],color=['red','green'])
```
